## SARIMA Model
### Introduction
Let $\{X_t\}$ is a nonstationary time series with trend and seasional period of *s*. If the differenced series

$$
\begin{align}
Y_t = (1-B)^d (1-B^s)^D X_t
\end{align}
$$
is stationary, then we are allowed to be going to consider the problem of finding an appropriate model of $\{Y_t\}$.

### Definition

If *d* and *D* are nonnegative integers and *B* is the backshift operator, then time series $\{X_t\}$ is a SARIMA$(p,d,q)(P,D,Q)_s$ process with *seasonal period* **s** if the differenced series $Y_t = (1 − B )^d (1 − B^s)^D  X_t$ is stationary and $\{X_t\}$ satisfies the following model:

$$
\begin{align*}
\varphi(B) \Phi(B^s)(1-B)^d(1-B^s)^D X_t &= \theta (B) \Theta(B^s) \epsilon_t
\end{align*}
$$

where:
- $\{\epsilon_t\}\sim WN(0,\sigma_{\epsilon}^2)$
- $\varphi_p\ne 0$
- $\theta_q \ne 0$.
- $\Phi_p\ne 0$
- $\Theta_q \ne 0$.
- $\varphi(z) = 1-\varphi_1 z - \varphi_2 z^2 - \cdots - \varphi_p z^p$
- $\theta (z) = 1+\theta_1 z+\theta_2 z^2 + \cdots + \theta_q z^q$
- $\Phi(z) = 1-\Phi_1 z - \Phi_2 z^2 - \cdots - \Phi_P z^P$
- $\Theta (z) = 1+\Theta_1 z+\Theta_2 z^2 + \cdots + \Theta_Q z^Q$

### Get in to the Simulation
We are going to simulate Catfish Sales data 

```python
def parser(s):
    return datetime.strptime(s, '%Y-%m-%d')

# reading the data and assigning to a variable
catfish_sales = pd.read_csv('catfish.csv', parse_dates=[0], index_col=0, date_parser=parser).squeeze(1)

#inferring the frequency of the data
catfish_sales = catfish_sales.asfreq(pd.infer_freq(catfish_sales.index))

# choosing sub data between the year of 1996-1999
start_date = datetime(1986,1,1)
#start_date = datetime(1996,1,1)
end_date = datetime(2000,1,1)
sub_catfish_sales = catfish_sales[start_date:end_date]
```

```python
len(sub_catfish_sales)
```

```python
# plotting
plt.figure(figsize=(10,4))
plt.plot(sub_catfish_sales)
plt.title('Catfish Sales in 1000s of Pounds', fontsize=14)
plt.ylabel('Sales', fontsize=14)
for year in range(start_date.year,end_date.year):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
```

#### Remove the trend
Since there is a trend as we can see in the plot, i.e. the mean was changing, then we are going to do a differencing one time and see the plot.

```python
first_diff = sub_catfish_sales.diff()[1:]

plt.figure(figsize=(10,4))
plt.plot(first_diff)
plt.title('Catfish Sales in 1000s of Pounds', fontsize=14)
plt.ylabel('Sales', fontsize=14)
for year in range(start_date.year,end_date.year):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
plt.axhline(0, color='k', linestyle='--', alpha=0.2)
```

#### Plotting ACF & PACF

```python
acf_pacf_fig(first_diff, lag = 40, both=True)
```

Based on ACF, we should start with a seasional MA process. Also, based on PACF, we should start with a seasional AR process.

#### Setting training and testing sets

```python
train_end = datetime(1999,7,1)
test_end = datetime(2000,1,1)

train_data = sub_catfish_sales[:train_end]
test_data = sub_catfish_sales[train_end + timedelta(days=1):test_end]
```

#### Fit the SARIMA Model

```python
pdq = (0,1,0)
PDQS = (1, 0, 1, 12)
# define model
model = SARIMAX(train_data, order=pdq, seasonal_order=PDQS)
```

```python
#fit the model
start = time()
model_fit = model.fit()
end = time()
print('Model Fitting Time:', end - start)
```

```python
#summary of the model
print(model_fit.summary())
```

```python
#get the predictions and residuals
predictions = model_fit.forecast(len(test_data))
predictions = pd.Series(predictions, index=test_data.index)
residuals = test_data - predictions
```

```python
plt.figure(figsize=(10,4))
plt.plot(residuals)
plt.axhline(0, linestyle='--', color='k')
plt.title('Residuals from SARIMA Model', fontsize=14)
plt.ylabel('Error', fontsize=14)
```

```python
plt.figure(figsize=(10,4))

plt.plot(sub_catfish_sales)
plt.plot(predictions)

plt.legend(('Data', 'Predictions'), fontsize=14)

plt.title('Catfish Sales in 1000s of Pounds', fontsize=14)
plt.ylabel('Production', fontsize=14)
for year in range(start_date.year,end_date.year):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
```

```python
print('MAPE:', round(np.mean(abs(residuals/test_data)),4))
```

```python
print('RMSE:', np.sqrt(np.mean(residuals**2)))
```

The prediction does follow the pattern of the data, but it is over predicting. It may be caused by predicting so many periods in the future. 

Instead of predicting several months, we are going to try predicting one month in the future, and then use it as the data training.

### Using the Rolling Forecast Origin

```python
rolling_predictions = test_data.copy()
for train_end in test_data.index:
    train_data = sub_catfish_sales[:train_end-timedelta(days=1)]
    model = SARIMAX(train_data, order=pdq, seasonal_order=PDQS)
    model_fit = model.fit()

    pred = model_fit.forecast()
    rolling_predictions[train_end] = pred
```

```python
rolling_residuals = test_data - rolling_predictions
```

```python
plt.figure(figsize=(10,4))
plt.plot(rolling_residuals)
plt.axhline(0, linestyle='--', color='k')
plt.title('Rolling Forecast Residuals from SARIMA Model', fontsize=14)
plt.ylabel('Error', fontsize=14)
```

```python
plt.figure(figsize=(10,4))

plt.plot(sub_catfish_sales)
plt.plot(rolling_predictions)

plt.legend(('Data', 'Predictions'), fontsize=14)

plt.title('Catfish Sales in 1000s of Pounds', fontsize=14)
plt.ylabel('Production', fontsize=14)
for year in range(start_date.year,end_date.year):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
```

```python
print('MAPE:', round(np.mean(abs(rolling_residuals/test_data)),4))
```

```python
print('RMSE:', np.sqrt(np.mean(rolling_residuals**2)))
```

Here we can compare that the errors in the later method is less (better) then the previous one.

### Detecting the Anomaly
#### Attemp 1

```python
plt.figure(figsize=(10,4))
plt.plot(sub_catfish_sales)
plt.title('Catfish Sales in 1000s of Pounds', fontsize=14)
plt.ylabel('Sales', fontsize=14)
for year in range(start_date.year,end_date.year):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
```

```python
rolling_deviations = pd.Series(dtype=float, index = sub_catfish_sales.index)
```

```python
for date in rolling_deviations.index:
    #get the window ending at this data point
    window = sub_catfish_sales.loc[:date]
    
    #get the deviation within this window
    rolling_deviations.loc[date] = window.std()
```

```python
#get the difference in deviation between one time point and the next
diff_rolling_deviations = rolling_deviations.diff()
diff_rolling_deviations = diff_rolling_deviations.dropna()
```

```python
plt.figure(figsize=(10,4))
plt.plot(diff_rolling_deviations)
plt.title('Deviation Differences', fontsize=14)
plt.ylabel('Sales', fontsize=14)
for year in range(start_date.year,end_date.year):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
```

#### Attempt 2: Seasonal Method

```python
month_deviations = sub_catfish_sales.groupby(lambda d: d.month).std()
```

```python
plt.figure(figsize=(10,4))
plt.plot(month_deviations)
plt.title('Deviation by Month', fontsize=14)
plt.ylabel('Sales', fontsize=14)
```

So, the anomaly occurs in a December

```python
december_data = sub_catfish_sales[sub_catfish_sales.index.month == 12]
```

```python
december_data
```

```python
min_dev = 9999999
curr_anomaly = None
for date in december_data.index:
    other_data = december_data[december_data.index != date]
    curr_dev = other_data.std()
    if curr_dev < min_dev:
        min_dev = curr_dev
        curr_anomaly = date
```

```python
curr_anomaly
```

### What should we do about the anomaly?
A: Simple Idea is to use mean of other months

```python
adjusted_data = sub_catfish_sales.copy()
adjusted_data.loc[curr_anomaly] = december_data[(december_data.index != curr_anomaly) & (december_data.index < test_data.index[0])].mean()
```

```python
plt.figure(figsize=(10,4))
plt.plot(sub_catfish_sales, color='firebrick', alpha=0.4)
plt.plot(adjusted_data)
plt.title('Catfish Sales in 1000s of Pounds', fontsize=14)
plt.ylabel('Sales', fontsize=14)
for year in range(start_date.year,end_date.year):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
plt.axvline(curr_anomaly, color='k', alpha=0.7)
```

### The Predictions

```python
train_end = datetime(1999,7,1)
test_end = datetime(2000,1,1)

test_data = adjusted_data[train_end + timedelta(days=1):test_end]
```

```python
rolling_predictions = test_data.copy()
for train_end in test_data.index:
    train_data = adjusted_data[:train_end-timedelta(days=1)]
    model = SARIMAX(train_data, order=pdq, seasonal_order=PDQS)
    model_fit = model.fit()

    pred = model_fit.forecast()
    rolling_predictions[train_end] = pred
```

```python
rolling_residuals = test_data - rolling_predictions
```

```python
plt.figure(figsize=(10,4))
plt.plot(rolling_residuals)
plt.axhline(0, linestyle='--', color='k')
plt.title('Rolling Forecast Residuals from SARIMA Model', fontsize=14)
plt.ylabel('Error', fontsize=14)
```

```python
plt.figure(figsize=(10,4))

plt.plot(sub_catfish_sales)
plt.plot(rolling_predictions)

plt.legend(('Data', 'Predictions'), fontsize=14)

plt.title('Catfish Sales in 1000s of Pounds', fontsize=14)
plt.ylabel('Production', fontsize=14)
for year in range(start_date.year,end_date.year):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', alpha=0.2)
```

```python
print('MAPE:', round(np.mean(abs(rolling_residuals/test_data)),4))
```

```python
print('RMSE:', np.sqrt(np.mean(rolling_residuals**2)))
```

The result is quite similar to the data without anomaly
